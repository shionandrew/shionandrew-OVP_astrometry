```python
import numpy as np
```

```python
def get_tot_error_theory(re_err,im_err):
    return np.sqrt(im_err**2+re_err**2)

def get_tot_error(d):
    im_err = np.nanstd(d.imag)
    re_err = np.nanstd(d.real)
    return get_tot_error_theory(re_err,im_err)



def get_mult_err(t,c):
    sig_RT = np.nanstd(t.real)
    sig_IT = np.nanstd(t.imag)
    
    sig_RC = np.nanstd(c.real)
    sig_IC = np.nanstd(c.imag)
    
    c_real = np.nanmean(c.real)
    t_real = np.nanmean(t.real)
    t_imag = np.nanmean(t.imag)
    c_imag = np.nanmean(c.imag)
    
    re_err = (sig_RT*c_real)**2 + (sig_RC*t_real)**2 + (t_imag*sig_IC)**2 + (c_imag*sig_IT)**2
    im_err = (sig_IT*c_real)**2 + (sig_RC*t_imag)**2 + (t_real*sig_IC)**2 + (c_imag*sig_RT)**2
    return get_tot_error_theory(np.sqrt(re_err),np.sqrt(im_err))

N=100
re_err=0.3
im_err=0.3

c_re_err=0.2
c_im_err=0.5

t = np.random.normal(0,re_err,N)+1j*np.random.normal(0,im_err,N)

c = np.random.normal(0,c_re_err,N)+1j*np.random.normal(0,c_im_err,N)


calibrated = t*c.conj()
print(get_tot_error(calibrated)/np.mean(np.abs(calibrated)))

print('theory:')
print(get_mult_err(t,c)/np.mean(np.abs(calibrated)))


calibrated = t/(c)#.conj()
print(get_tot_error(calibrated)/np.mean(np.abs(calibrated)))


calibrated = t*c.conj()/np.abs(c)
print(get_tot_error(calibrated)/np.mean(np.abs(calibrated)))



```
